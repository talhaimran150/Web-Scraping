#Question No 2

##Part a

```python
def singleDeterminer(string):
    patternn = r'\b(?:a|an|the)\b'
    match = re.search(patternn, string)
    result = "{}: {}".format(string, match.group()) if match else "{}: No match".format(string)
    return result

testPartA = ["The apple","The" "an", "A","a"]
for Iterator in testPartA:
    print(singleDeterminer(Iterator))
```

##Part b

```python
def arithmeticExpression(string):
    patternnn = r'\b(?:\d+(?:[+*]\d+)+)\b'
    match = re.search(patternnn, string)
    result = "{}: {}".format(string, match.group()) if match else "{}: No match".format(string)
    return result

testPartB = ["8*2-1", "4*2", "11+22", "xyz","4*6+2"]
for Iterator in testPartB:
    print(arithmeticExpression(Iterator))
```

```python
def utilityFunction(url):
    htmlContent = request.urlopen(url).read().decode('utf8')
    print("HTML Content:")
    print(htmlContent)
    soup = BeautifulSoup(htmlContent, 'html.parser')
    textContent = soup.get_text()
    print("Text Content:")
    print(textContent)

URL = 'https://www.dsu.edu.pk/contact-us/'
utilityFunction(URL)
```

#Question No 4

```python
url = "https://www.dsu.edu.pk/contact-us/"
response = requests.get(url)
text = response.text
print(text)
```

```python
tokenizer = RegexpTokenizer(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b|\b\d{3}[-.\s]?\d{3}[-.\s]?\d{4}\b')
tokens = tokenizer.tokenize(text)
print(tokens)
```

```python
phoneNumber = [token for token in tokens if re.match(r'\b\d{3}[-.\s]?\d{3}[-.\s]?\d{4}\b', token)]
print("Phone Number")
print(phoneNumber)
email = [token for token in tokens if re.match(r'\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b', token)]
print("Email Address")
print(email)
```

#Question No 5

## Porter Stemmer

```python
porterStemmer = PorterStemmer()
normPorterStemmer = [porterStemmer.stem(word) for word in tokens]
print(normPorterStemmer)
```

## Lancaster Stemmer

```python
lancasterStemmer = LancasterStemmer()
normLancasterStemmer = [lancasterStemmer.stem(word) for word in tokens]
print(normLancasterStemmer)
```
